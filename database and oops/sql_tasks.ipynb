```python
import mysql.connector
from mysql.connector import Error

class MySQLDatabase:
    def __init__(self, host, port, user, password, database=None):
        self.host = host
        self.port = port
        self.user = user
        self.password = password
        self.database = database
        self.connection = None

    def connect(self):
        """Establishes a connection to the MySQL server."""
        try:
            self.connection = mysql.connector.connect(
                host=self.host,
                port=self.port,
                user=self.user,
                password=self.password
            )
            if self.connection.is_connected():
                print("Successfully connected to the MySQL server.")
                if self.database:
                    self.use_database(self.database)
        except Error as e:
            print(f"Error: {e}")

    def create_database(self, database_name):
        """Creates a new database if it does not already exist."""
        cursor = None
        try:
            cursor = self.connection.cursor()
            cursor.execute(f"SHOW DATABASES LIKE '{database_name}'")
            result = cursor.fetchone()
            if result:
                print(f"Database '{database_name}' already exists.")
            else:
                cursor.execute(f"CREATE DATABASE {database_name}")
                print(f"Database '{database_name}' created successfully.")
        except Error as e:
            print(f"Error: {e}")
        finally:
            if cursor:
                cursor.close()

    def use_database(self, database_name):
        """Selects the database to use."""
        cursor = None
        try:
            cursor = self.connection.cursor()
            cursor.execute(f"USE {database_name}")
            print(f"Switched to database '{database_name}'.")
        except Error as e:
            print(f"Error: {e}")
        finally:
            if cursor:
                cursor.close()

    def close(self):
        """Closes the connection to the MySQL server."""
        if self.connection and self.connection.is_connected():
            self.connection.close()
            print("Connection closed.")

    def insert(self, table, columns, values):
        """Inserts data into a table."""
        cursor = None
        try:
            cursor = self.connection.cursor()
            columns_str = ', '.join(columns)
            placeholders = ', '.join(['%s'] * len(values))
            query = f"INSERT INTO {table} ({columns_str}) VALUES ({placeholders})"
            cursor.execute(query, values)
            self.connection.commit()
            print("Record inserted successfully.")
        except Error as e:
            print(f"Error: {e}")
        finally:
            if cursor:
                cursor.close()

    def update(self, table, set_values, where_clause):
        """Updates records in a table."""
        cursor = None
        try:
            cursor = self.connection.cursor()
            set_clause = ', '.join([f"{col} = %s" for col in set_values.keys()])
            query = f"UPDATE {table} SET {set_clause} WHERE {where_clause}"
            cursor.execute(query, list(set_values.values()))
            self.connection.commit()
            print("Record updated successfully.")
        except Error as e:
            print(f"Error: {e}")
        finally:
            if cursor:
                cursor.close()

    def delete(self, table, where_clause):
        """Deletes records from a table."""
        cursor = None
        try:
            cursor = self.connection.cursor()
            query = f"DELETE FROM {table} WHERE {where_clause}"
            cursor.execute(query)
            self.connection.commit()
            print("Record(s) deleted successfully.")
        except Error as e:
            print(f"Error: {e}")
        finally:
            if cursor:
                cursor.close()

    def fetch_all(self, query, params=None):
        """Fetches all records from a query."""
        cursor = None
        try:
            cursor = self.connection.cursor(dictionary=True)
            cursor.execute(query, params)
            result = cursor.fetchall()
            return result
        except Error as e:
            print(f"Error: {e}")
            return []
        finally:
            if cursor:
                cursor.close()

    def fetch_one(self, query, params=None):
        """Fetches a single record from a query."""
        cursor = None
        try:
            cursor = self.connection.cursor(dictionary=True)
            cursor.execute(query, params)
            result = cursor.fetchone()
            return result
        except Error as e:
            print(f"Error: {e}")
            return None
        finally:
            if cursor:
                cursor.close()
```
